```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ptspath = 'C:/Users/tingh/5934 compute statistic python/HW07pts.txt'
pts = np.loadtxt(ptspath,delimiter=',')
wtspath = 'C:/Users/tingh/5934 compute statistic python/HW07wts.txt'
wts = np.loadtxt(wtspath,delimiter=',')
m,n = np.shape(wts)
```

```python
#Problem 1a
plt.scatter(pts[:,0],pts[:,1])
for i in range(m):
    for j in range(n):
        if wts[i,j] < 1000:
            plt.arrow(pts[i][0],pts[i][1],(pts[j][0] - pts[i][0]), (pts[j][1] - pts[i][1]))
```

```python
#Problem 1b
for i in range(m):
    for j in range(n):
        if wts[i,j] >= 1000:
            wts[i,j] = 0

import networkx as nw
G = nw.Graph(wts)
T = nw.minimum_spanning_tree(G, algorithm="prim")
edges = list(T.edges)

plt.scatter(pts[:,0],pts[:,1])
for i in range(len(edges)):
    start = edges[i][0]
    stop = edges[i][1]
    plt.arrow(pts[start][0],pts[start][1],(pts[stop][0] - pts[start][0]), (pts[stop][1] - pts[start][1]))
```

```python
#Problem 1c
pred, pred_ = nw.floyd_warshall_predecessor_and_distance(G)
plt.scatter(pts[:,0],pts[:,1])
for i in range(m):
    for j in range(n):
        if i != j:
            s_path = nw.reconstruct_path(i, j, pred)
            nextp = s_path[1]
            plt.arrow(pts[i][0],pts[i][1],(pts[nextp][0] - pts[i][0]), (pts[nextp][1] - pts[i][1]))
    
```

```python
#Problem 1d
shortestpath = nw.reconstruct_path(82, 68, pred)

plt.scatter(pts[:,0],pts[:,1])
for i in range(len(shortestpath)-1):
    start = shortestpath[i]
    stop = shortestpath[i+1]
    plt.arrow(pts[start][0],pts[start][1],(pts[stop][0] - pts[start][0]), (pts[stop][1] - pts[start][1]),head_width = 4)
    
```
