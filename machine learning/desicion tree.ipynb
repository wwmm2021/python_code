```python
######## hw1  ###########################
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
# from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
##################   problem a   #########################
# data import
train_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_train.data")
valid_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_valid.data")
test_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_test.data")
valid_labels = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_valid.labels")
train_labels = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_train.labels")
```

```python
# built decision tree
x=[]
y=[]
y_train = []
b=()
c=()
for i in range(1,13,1):
    clf = DecisionTreeClassifier(criterion='entropy',max_depth = i)
    clf.fit(train_data,train_labels)
    y_ = clf.predict(valid_data)
    y_train_ = clf.predict(train_data)
    b =1- accuracy_score(valid_labels,y_)
    c =1- accuracy_score(train_labels,y_train_)
    # tree.plot_tree(clf,filled=True)
    x.append(i)
    y.append(b)
    y_train.append(c)
```

```python
# plot the figure
plt.plot(x,y,'r.-',label = 'test')
plt.plot(x,y_train,'b.-',label = 'train')
plt.title('tree_depth VS predict Error')
plt.ylabel('Error')
plt.xlabel('tree-depth')
print('******* result for problem a ********')
print('the min error is:', min(y))
#################################print('the depth is: ',argmin(y))
print('the min error tree-depth is 5')
plt.legend()
```

```python
###########################    problem b    ########################################
# data import
test_data = pd.read_csv('D:/2020fall/5635 machine learning/5635 hw/hw1/wilt_test.csv',header=None,sep=",")
test_labels = pd.read_table('D:/2020fall/5635 machine learning/5635 hw/hw1/wilt_test.labels',header=None,sep=",")
train_data = pd.read_csv('D:/2020fall/5635 machine learning/5635 hw/hw1/wilt_train.csv', header=None,sep =",")
train_labels = pd.read_table('D:/2020fall/5635 machine learning/5635 hw/hw1/wilt_train.labels',header=None,sep=",")
```

```python
##############   problem c  ################
# data import
train_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_train.data")
valid_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_valid.data")
test_data = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_test.data")
valid_labels = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_valid.labels")
train_labels = np.loadtxt(r"D:\2020fall\5635 machine learning\5635 hw\hw1\madelon_train.labels")
```

```python
from sklearn.ensemble import RandomForestClassifier
# built decision tree
x=[]
y=[]
b=()
y_train = []
b=()
c=()
for i in [3,10,30,100,300]:
    clf = RandomForestClassifier(n_estimators=i,max_features=22)
    clf.fit(train_data,train_labels)
    y_ = clf.predict(valid_data)
    y_train_ = clf.predict(train_data)
    b =1- accuracy_score(valid_labels,y_)
    c =1- accuracy_score(train_labels,y_train_)
    x.append(i)
    y.append(b)
    y_train.append(c)
plt.figure()
plt.plot(x,y,'r.-',label = 'test')
plt.plot(x,y_train,'b.-',label = 'train')
plt.title('under max_feature = 22 K_trees VS predict Error')
plt.ylabel('Error')
plt.xlabel('K_tree')
print('******* result for problem c ********')
print('different max_feature:',x)
print('...........test Error:',y)
print('..........train Error:',y_train)
plt.legend()
```
